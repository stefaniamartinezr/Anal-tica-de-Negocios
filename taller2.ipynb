Una entidad del sector salud (Las Americas) desea implementar un modelo de árbol de decisión con el fin de predecir si un paciente desarrollará una enfermedad cardíaca. El objetivo es apoyar el diagnóstico preventivo a partir de variables clínicas básicas, permitiendo identificar grupos de riesgo de manera temprana.

Las variables son:

*Age (Edad): Edad del paciente en años, factor clave en el riesgo cardiovascular.

*Systolic (Presión Sistólica): Presión arterial cuando el corazón se contrae.

*Diastolic (Presión Diastólica): Presión arterial cuando el corazón está en reposo.

*BMI (Índice de Masa Corporal): Relación entre peso y estatura, asociada al riesgo de enfermedad cardíaca.

0. se procee con la carga de las librerias de trabajo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_text
```

1. se procede con la carga de los datos de trabajo

```python
nxl='/content/2. HeartDisease.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XD=XDB.iloc[:,[1,2,3,4]]
yd=XDB.iloc[:,0]
display(XD)
display(yd)
```

2. Se procede con la implementación del modelo del árbol


```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=4)
mar.fit(XD,yd)    #Aqui el modelo busca relación entrada-salida

ydp=mar.predict(XD)    #Esto es lo que prónostica el modelo

#Se construye la matriz de consfusión
cm=confusion_matrix(yd,ydp)
display(cm)
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]
#Metricas de Desempeño
Ex=(VP + VN)/len(XD)    #1. Exactitud: Comportamiento General
print("La exactitud es:" ,Ex)
Sen=VP/(VP+FN)     #2. Sensibilidad: Como se comporta pronosticando PreAprobados
print("La sensibilidad es: ",Sen)
Spe=VN/(VN+FP)  #3. Especificidad: Como se comporta frente a los negativos
print("La especificidad es:",Spe)
Pre=VP/(VP+FP)  #4. Precisión: Como se comporta pronosticando PreAprobados
print("La Precición es: ",Pre)
PreNeg=VN/(VN+FN) #5. Precisión Negativa: Como se comporta pronósticando negativos
print("La Predicción Negativa es:",PreNeg)
```

3. Despliegue del Árbol de *Decisión*

```python
from sklearn.tree import export_graphviz     #Exporta los datos a un gráfico
from pydotplus import graph_from_dot_data    #Es un graficador

vs=["Age", "Systolic","Diastolic", "BMI"]   #Titulos del Árbol
dot_data=export_graphviz(mar,feature_names=vs) #Exportar de números a gráfico
graph=graph_from_dot_data(dot_data)
graph.write_png('Arbol.png')
```

4. Aqui lo muestra la probabilidad de enfermedad según las variables

```python
#Leer hoja 2
X2DB = pd.read_excel(nxl, sheet_name=1)   # hoja 2 del Excel
X2 = X2DB[["Age", "Systolic", "Diastolic", "BMI"]]  # mismas columnas

# Aqui se Predice con el modelo entrenado
pred_hoja2 = mar.predict(X2)

# la predicción a la tabla
X2DB["Prediccion_HeartDisease"] = pred_hoja2

display(X2DB)
```

Se identifican nodos puros (GINI = 0), un ejemplo coresponde:

Edad > 69.5, lo que conduce a un nodo puro con 2 observaciones correctamente clasificadas (2/0).

Al analizar el árbol de decisión se identifican en total cinco nodos puros. En tres de estos nodos se observa que los valores son samples = 12, value = [0,12] (no cumple systolic <= 123), samples = 35, value = [0,35] (no cumple Diastolic <= 85.5)y samples = 2, value = [0,2] (si cumple diastolic <= 60.5). Esto significa que en esos grupos específicos todos los pacientes (12, 35 y 2 respectivamente) presentan enfermedad, ya que el primer número del “value” corresponde a los pacientes sin enfermedad (0) y en el segundo a los que sí la tienen enfermedad (1). En estos casos, el modelo no presenta ningún error dentro del nodo, por eso el gini es 0 (puro) y la probabilidad de enfermedad es del 100%.

Por otro lado, también se identifican dos nodos puros relacionados  a laausencia de enfermedad: uno con samples = 2, value = [2,0] (no cumple edad <= 69.5)y otro con samples = 1, value = [1,0] (si cumple Diastolic <= 70). En estos nodos todos los pacientes clasificados no presentan enfermedad, ya que“samples” indica cuántos pacientes llegaron a ese nodo y “value” muestra cómo se distribuyen entre las dos clases (sin enfermedad 0 y con enfermedad 1).

Los nodos puros (GINI = 0) indican reglas del negocio extremadamente claras, mientras que los nodos extremos (con porcentajes del 98%–99%) representan zonas del modelo donde la probabilidad de clasificación es muy alta, aunque no totalmente perfecta.
