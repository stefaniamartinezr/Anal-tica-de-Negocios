#**Caso de Estudio**
En este documento se desarrolla y analiza un modelo Naive Bayes para la preAprobación de créditos en una entidad financiera utilizando las variables socioeconomicas de un solicitante de credito. De acuerdo con lo anterior las variables socioeconómicas son:
**Edad**: Es la edad que posee el solicitante de credito al momento de solicitar un crédito.
**Ingresos**: Ingresos recibidos por una persona como resultado de una actividad comercial.
**Egresos**: Son los gastos mensuales del solicitante. Gastos de personas a cargo, hijos y servivios públicos.
**Monto (EAD)**: Indica el dinero que una persona solicita a una entidad finciera como crédito.


0. Se preocede con la carga de las librerias de trabajo

```python
import numpy as np #Libreria numérica por excelencia.
import pandas as pd #Libreria para la comunicación con archivos de excel

#Librerias especificas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos BaseAAA

```python
nxl='/content/1. SolicitantesCrédito(USD) (1).xlsx'
XDB=pd.read_excel(nxl,sheet_name=1)
#XDB.head(100) #Es para mostrar la base de datos cargada


#Seleccionamos variables de trabajo
XD=XDB[['Edad', 'Ingresos', 'Egresos', 'Monto (EAD)']] #Variables de Trabajo
XD.head(10)
yd=XDB[['PreApr']]  #Esta la variable de decisión
yd.head()
```

2. Implementamos Modelo Naive Bayes

```python
mnb=GaussianNB()
mnb.fit(XD,yd)  #Ajustar Variables Entrada - Salida

#Mostramos las medias de las variables
u=mnb.theta_
sigma=mnb.var_; sigma=np.sqrt(sigma)
print("'Edad','Ingresos','Egresos','Monto (EAD)'")
print(u)
print("Las desviaciones son:")
print(sigma)
```

**Análisis de Resultados**
De los resultados arrojados por el modelo, podemos observar que la categoría de PreAprobados posee ingresos superiores a las categorías de PreNegados, esto indica claramente la predominancia de la variable para aprobar créditos. Es importante mencionar que la edad es mayor para los preAprobados que para los PreNegados, lo que indica que el banco busca personas con mayor experiencia créditicia. Podemos decir igualmente, que los PreAprobados están solicitantes montos superiores a los de la categoría de PreNegados, esto es un indicador de la capacidad de este grupo para hacer frente a un crédito.


3. Se procede con la evaluación del modelo. Para la evaluación de este tipo de modelo se utiliza la matriz de confusión

```python
ydp=mnb.predict(XD) #Esto es lo que el modelo aprende - ydp de pronóstico
cm=confusion_matrix(yd,ydp)
print(cm)

#Se determinan las metricas de la matriz de confusión
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1];TDatos=len(XDB)

#1. Exactitud: Funcionamiento general del modelo
Ex=(VP+VN)/TDatos
print("Exactitud: ",Ex)

#2. Tasa Error: %Fallos del Modelo #Si apartir del 75% el modelo es bueno, es decir, de 4 datos, le atino a 3 y solo tuvo error en 1
TEr=(FP+FN)/TDatos
print("Tasa Error: ",TEr)

#3. Sensibilidad: Como se comporto conn respecto a los positivos solamente
Se=VP/(VP+FN)
print("Sensibilidad: ", Se)

#4. Especificidad: Como se comporta pronosticando negativos
Es=VN/(VN+FP)
print("Especificidad: ", Es)

#5. Precisión: Es una versión de como se comporta el modelo frente a los positivos solamente
Pr=VP/(VP+FP)
print("Precisión: ", Pr)

#6. Predicción Negativa: Como funciona el modelo pronósticando créditos PreNegados
PrN=VN/(VN+FN)
print("Predicción Negativa: ", PrN)

```

**Análisis de Resultados**
De acuerdo con los r4esultados, podemos observar de manera general que el modelo alcanzó una aexactitud por encima del 75% en promedio, lo que indica el buen comportamiento del modelo frente a la clasificación de los solicitantes de crédito frente a la PreApr y la PreNegación. Se destaca la precisión con un 80.9%, lo que indica el buen funcionamiento del modelo frente a la PreAprobación de créditos. Igualmente se puede observar el buen comportamineto del modelo frente a la PreNegación de créditos, tal y como lo demuestra el indice de especificidad del modelo, el cual estuvo por encima del 84% en promedio.
Es importante mencionar que la excatitud queda por debajo del 80% ya ue está métrica se ve afectada por los errores en la clasificación de solicitantes de crédito, lo que se evidencia a tráves de la Sensibilidad y la Predicción Negativa.
